# Desafío 1
Generar 10 números aleatorios entre 100 y 200 y calcular su media.

```python
from random import randint
from statistics import mean

def generaLista():
    lista = []
    for i in range(10):
        lista.append(randint(100,200))
    return lista

lista = generaLista()
print(lista)
print("La media es",mean(lista))
```
