# Condicionales `If`

```python
if <condición 1>:
    <código>
elif <condición 2>:
    <código>
elif <condición 3>:
    <código>
else:
    <código>
```



````python
edad = 20
if edad >= 18:
    print('Eres mayor de edad.')
elif edad >= 13:
    print('Eres un adolescente.')
else:
    print('Eres un niño.')
````

````python
ciudades =["Roma", "Londres", "París", "Madrid", "Berlín"]
city = "Austin"

if city in ciudades:
    print(f"{city} si se encuentra en la lista de ciudades.")
else:
    print(f'{city} no se encuentra en la lista de ciudades.')
````
