```python
x = 5
0 < x and x < 10    # True
```

```python
0 < x < 10          # True
```

## Condición lógica en un condicional

```python
a = 7
b = 9

if a == b:
    print("Si son iguales")

if a != b:
    print("No son iguales")
```
